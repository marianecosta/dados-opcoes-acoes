```python
import pandas as pd
import requests 
```

```python
def listar_opcoes(ativo_obj, vencimento):
  url = f'https://opcoes.net.br/listaopcoes/completa?idAcao={ativo_obj}&listarVencimentos=false&cotacoes=true&vencimentos={vencimento}'
  r = requests.get(url).json()
  l = [[ativo_obj, vencimento, i[0].split('_')[0], i[2], i[3], i[4], i[5], i[8], i[9], i[10]] for i in r['data']['cotacoesOpcoes']]
  return pd.DataFrame(l, columns=['ativo_obj', 'Vencimento', 'Ticker', 'Tipo', 'Modelo', 'A/I/OTM','Strike', 'Preço', 'N. de negócios', 'Volume Neg.'])
 
```

```python
def listar_tudo(ativo_obj):
  url=f'https://opcoes.net.br/listaopcoes/completa?idAcao={ativo_obj}&listarVencimentos=true&cotacoes=true'
  r = requests.get(url).json()
  vencimentos = [i['value'] for i in r['data']['vencimentos']]
  df = pd.concat([listar_opcoes(ativo_obj, vencimento) for vencimento in vencimentos])
  return df
```

```python
listar_tudo('PETR4')
```

```python
aux = listar_tudo('PETR4')

# Procura opções com vencimento em 2022-07-15 e com número de negócios maior que 700
aux[(aux['Vencimento'] == '2022-07-15') & (aux['N. de negócios'] > 700)]
```
